```python

from transformers import pipeline

# Load a text-generation model fine-tuned on Indian legal data
qa_pipeline = pipeline("text-generation", model="bigcode/legal-t5-small")

def generate_explanation(section_text):
    prompt = f"Explain the following legal term in simple words: {section_text}"
    response = qa_pipeline(prompt, max_length=100)[0]["generated_text"]
    return response

# Example
print(generate_explanation("Section 304B IPC"))
```

```python
import pandas as pd  

# Load both Parquet files  
df1 = pd.read_parquet("legal1.parquet")  
df2 = pd.read_parquet("legal2.parquet")  

# Merge them using concat  
df_merged = pd.concat([df1, df2], ignore_index=True)  

# Save the merged file  
df_merged.to_parquet("merged_legal.parquet", index=False)

print("Merged Parquet file saved successfully!")
```
